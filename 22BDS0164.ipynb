```python
#Data Cleaning
df['dos'] = df['dos'].astype(str)
df['dos'].str.isnumeric().value_counts()
```

```python
airq = df['airq'].loc[df['airq'] != '?']
pmean = airq.astype(int).mean()
df['airq'] = df['airq'].replace('?', pmean).astype(int)
df['airq'].head(10)

```

```python
airq = df['airq'].loc[df['airq'] != '?']
aqmean = airq.astype(int).mean()
df['airq'] = df['airq'].replace('?', aqmean).astype(int)
df['airq'].head()

```
